 Pie Chart Sederhana

```python
import matplotlib.pyplot as plt

# Data partisipasi siswa
kegiatan = ['Olahraga', 'Seni', 'Musik', 'Debat']
persentase = [30, 20, 25, 25]

# Membuat pie chart
plt.pie(persentase, labels=kegiatan, autopct='%1.1f%%', startangle=140)
plt.title('Persentase Partisipasi Siswa dalam Kegiatan Ekstrakurikuler')
plt.show()
```

Pie Chart dengan Matplotlib

```python
import matplotlib.pyplot as plt

# Data partisipasi siswa
kegiatan = ['Olahraga', 'Seni', 'Musik', 'Debat']
persentase = [30, 20, 25, 25]

# Warna khusus
colors = ['gold', 'lightblue', 'lightcoral', 'limegreen']

# Membuat pie chart dengan warna
plt.pie(persentase, labels=kegiatan, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Persentase Partisipasi Siswa dengan Warna Kustom')
plt.show()
```

Kustomisasi Pie Chart


```python
import matplotlib.pyplot as plt

# Data partisipasi siswa
kegiatan = ['Olahraga', 'Seni', 'Musik', 'Debat']
persentase = [30, 20, 25, 25]

# Warna khusus
colors = ['gold', 'lightblue', 'lightcoral', 'limegreen']

explode = [0.1, 0, 0, 0] # Bagian pertama (Olahraga) sedikit dikeluarkan

plt.pie(persentase, labels=kegiatan, colors=colors, autopct='%1.1f%%', startangle=140, explode=explode, shadow=True)
plt.title('Persentase Partisipasi Siswa')
plt.legend(kegiatan, loc='best')
plt.show()
```

Multiple Horizontal Bar Chart


```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Data kelompok 1
kegiatan_1 = ['Olahraga', 'Seni']
persentase_1 = [40, 60]

# Data kelompok 2
kegiatan_2 = ['Musik', 'Debat']
persentase_2 = [50, 50]

# Pie chart untuk kelompok 1
axs[0].pie(persentase_1, labels=kegiatan_1, autopct='%1.1f%%', startangle=140)
axs[0].set_title('Partisipasi: Kelompok 1')

# Pie chart untuk kelompok 2
axs[1].pie(persentase_2, labels=kegiatan_2, autopct='%1.1f%%', startangle=140)
axs[1].set_title('Partisipasi: Kelompok 2')

plt.suptitle('Perbandingan Partisipasi Siswa')
plt.show()
```
